```python
#loop  control stattment
#break :
    #used to exit from loop
i=10
while(i>1):
    
    if(i==5):
        break;
    print(i)
    i=i-1
```

```python
#continue
i=10
while(i>1):
    if(i==5):
        i=i-1
        continue
    print(i)
    i=i-1
```

```python
for i in range(2, 101):   # numbers from 2 to 100
    d = 0                 # reset divisor count for each i
    for n in range(1, i+1):   # check divisors of i
        if i % n == 0:
            d += 1
    if d == 2:            # prime has exactly 2 divisors
        print(i)
```
